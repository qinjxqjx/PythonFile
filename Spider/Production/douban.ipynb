```python
from bs4 import BeautifulSoup
import requests

content = requests.get('https://book.douban.com/').text
soup = BeautifulSoup(content,'lxml')
dict = {}

for link in soup.find_all('li',{"class":""}):
    for ref in link.find_all('div', {"class":"cover"}):
        print(ref.a.get("href")) 
    for name in link.find_all('div',{"class":"more-meta"}):
        print(name.h4.get_text())
```
